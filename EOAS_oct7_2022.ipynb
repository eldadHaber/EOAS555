```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.autograd import grad

import matplotlib.pyplot as plt
import numpy as np
```

```python
def firstFun(x, theta):
  r = x-theta
  return 0.5*torch.sum(r**2)

def banana(x, theta=[1,100]):
  f = (theta[0]-x[0])**2 + theta[1]*(x[1]-x[0]**2)**2
  return f
```

```python
theta = torch.randn(10)
x = torch.randn(10)
loss = firstFun(x, theta)
print(loss)
```

```python
x = torch.tensor([3,3.0])
theta = torch.tensor([1,100.0])
y = banana(x, theta)
print(y)
```

```python
def SD(fun, x, theta, maxiter, lr):

  for k in range(maxiter):

    xtemp = x.clone().detach()
    xtemp.requires_grad = True
    f = fun(xtemp, theta)
    df = grad(f,xtemp)[0]
    ng = torch.mean(torch.abs(df))

    # line search
    with torch.no_grad():
      cnt = 1
      mu = lr
      ftry = f
      while ftry >= f:
        if cnt > 8:
          print('line search fail')
          return x
        xtry = x - mu * df
        ftry = fun(xtry, theta)
        print('%d %d   %3.2e.   %3.2e'%(k, cnt, f.item(), ftry.item()))
        mu = mu/2
        cnt += 1
 

    x = xtry
    print('%d   %3.2e.   %3.2e'%(k, f.item(), ng.item()))
  return x
```

```python
xfinal = SD(banana, x, theta, 1000, 0.001)
```
